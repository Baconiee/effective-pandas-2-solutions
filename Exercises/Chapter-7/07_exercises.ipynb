```python
import pandas as pd

ser = pd.Series([1, 2, 3], name="my_series", dtype="int64[pyarrow]")
```

```python
# nbytes attribute
ser.nbytes
```

```python
# dtype attribute
ser.dtype
```

```python
# mean attribute
ser.mean()
```

```python
# values attribute
ser.values
```

```python
# iloc attribute
ser.iloc[2]
```

2.

```python
str_ser = pd.Series(["apple", "banana", "cherry"], name="my_str_series", dtype="string[pyarrow]")
# This also comprises private attribues(dunder methods)
len(dir(str_ser.str))
```

```python
# Without dunder methods
public_attributes = [attr for attr in dir(str_ser.str) if not attr.startswith("__")]
len(public_attributes)
```

```python
# upper attribute
str_ser.str.upper()
```

```python
# lower attribute
str_ser.str.lower()
```

```python
# title attribute
str_ser.str.title()
```
