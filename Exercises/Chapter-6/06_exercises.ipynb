```python
import pandas as pd

# Create a series with the temperature values for the last seven days
temp_ser = pd.Series([12, 14, 16, 14, 12, 13, 13], name='Temperature')

# Filter the series to include only the days where the temperature was above the average
mask = temp_ser >= temp_ser.mean()
temp_ser[mask]
```

```python
import pandas as pd

fav_colors_ser = pd.Series(['blue', 'green', 'burgundy', 'brown'], dtype='category')
fav_colors_ser
```

```python
fav_colors_ser.cat.ordered
```

Extra: Make categorical Series ordered.


```python
cat_type = pd.CategoricalDtype(categories=['blue', 'brown', 'burgundy', 'green'],
                               ordered=True)

fav_colors_ser = fav_colors_ser.astype(cat_type)
fav_colors_ser.cat.ordered
```
