3.

The proper type is 'string' (or specifically 'string[pyarrow]'). We avoid casting to str or object because those use legacy NumPy storage, which is less efficient and doesn't handle missing values as cleanly as the dedicated String Extension dtypes.

4.

```python
str_ser = pd.Series(['d', 'c', 'a', 'b', 'b', 'd'], dtype='string[pyarrow]')
```

```python
# Question 5
str_ser.memory_usage(deep=True)
```

```python
# Method 1 (Use this for simple cases)
str_ser = str_ser.astype('category').cat.as_ordered()
```

```python
# Question 5
str_ser.memory_usage(deep=True)
```

```python
# Method 2(Use this when u need a better control)
values = pd.Series(sorted(set(str_ser)))
str_ser_type = pd.CategoricalDtype(categories=values, ordered=True)
str_ser.astype(str_ser_type)
```
